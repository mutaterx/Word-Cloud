Objective
- Using a word cloud we will identify the freqeuency of certain words that appeared in book titles

```python
import pandas as pd 
    
pd.pandas.set_option('display.max_columns', None)
df = pd.read_csv("downloads/manga.csv")
df 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
```

# Combining the strings in the text column 

```python
# creating the text variable
# Text will join all the titles together so that python can analyse every word for the word cloud
text = " ".join(title for title in df.title)
```

# Plotting the word cloud

Understanding "collocations=False"
- Collocations in python is a sequence of words that occurs together unusually often.
- Here ***collocations=False*** to ensure that the word cloud doesn’t appear as if it contains any duplicate words. <br>
Otherwise, you may see for example ‘one’, ‘piece’ and ‘one piece’ as a collocation in the word cloud, giving an impression that words have been duplicated. <br>

```python
# Creating word_cloud with text as argument in .generate() method
word_cloud = WordCloud(collocations = False, background_color = 'white', width = 2048, height = 1080).generate(text)

#Saving the image
word_cloud.to_file('got.png')
```

```python
# Display the generated Word Cloud
plt.imshow(word_cloud)

# Axis "off", otherwise the plot will have a yand x axis
plt.axis("off")

# Plot the word cloud
plt.show()
```

From the word cloud we can see that "shoujo" is quite frequent in recurrent book titles. In the context of the exercise we can suppose that customers are quite interested in shoujo mangas (romantic mangas). Other frequent word like Love and Kiss, contribute to this idea. <br>
Additionally, we have the word "Isekai" that is apparent. Isekai is a type of manga genre that is quite popular among readers.

# Choosing Groupby 

Using Group by will give us a different result. Here we see the amount of times each manga appeared in the dataframe.

```python
# Counting the amount of times a title appears in the column "title"
groupby_df = df.groupby(["title"]).sum()

# Sorting the count from DESC order 
sort = groupby_df.sort_values('manga_id', ascending=False)

# Display the new group by table
sort
```
